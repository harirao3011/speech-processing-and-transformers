```python
from transformers import pipeline
```

```python
summarize = pipeline('summarization')
```

```python
try:
    text = input('Enter the text: ')
    print('You entered:', text)
except KeyboardInterrupt:
    print('\nInput was interrupted by the user.')
```

```python
result = summarize(text,max_length = 100, min_length = 30, do_sample = False)
print(result[0]['summary_text'])
```

```python
from googletrans import Translator
translator = Translator()
english_text = input("Enter text in English: ")
target_lang = input("Enter target language code (e.g., 'fr' for French, 'es' for Spanish): ")

translated = translator.translate(english_text, src='en', dest=target_lang)

print(f"\nTranslated text in {target_lang}: {translated.text}")
```
